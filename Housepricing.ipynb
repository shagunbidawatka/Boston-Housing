```python
#importing the required libraries
import numpy as np            #arrays
import pandas as pd             #data

import matplotlib as mpl          #graph work
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML
```

```python
#load dataset
boston=load_boston()

print(boston.DESCR)
```

```python
#put data in pandas Data frames
features=pd.DataFrame(boston.data,columns=boston.feature_names)
features
```

```python
#getting our target values i.e. prices/output
target=pd.DataFrame(boston.target,columns=['target'])
target
```

```python
max(target['target'])
```

```python
min(target['target'])
```

```python
#concatenate features and target into single data frame
#axis=1 makes it concatenate column wise
df=pd.concat([features,target],axis=1)
df
```

```python
df.describe().round(decimals=2)
```

```python
#correlation
corr=df.corr(method='pearson')

#take absolute values of correlation
corrs=[abs(corr[attr]['target']) for attr in list(features)]

#make list of pairs 
l=list(zip(corrs,list(features)))

#sort desc
#with correlation value as the key for sorting
l.sort(key=lambda x:x[0],reverse=True)

#unzip pairs to 2 lists
#zip(*l) takes list [[a,b,c],[d,e,f],[g,h,k]]
#and returns [[a,d,g],[b,e,h],[c,f,k]]
corrs,labels=list(zip((*l)))

#plot correlation w.r.t target var
index=np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)
plt.xlabel('Attributes')
plt.ylabel('correlation with target')
plt.xticks(index,labels)
plt.show()
```

```python
X=df['LSTAT'].values
Y=df['target'].values
```

```python
#before normalization
print(Y[:5])
```

```python
x_scaler=MinMaxScaler()
x=x_scaler.fit_transform(X.reshape(-1,1))
X=x[:-1]
y_scaler=MinMaxScaler()
y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=y[:-1]
```

```python
#after normalization
Y[:5]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
lin_model.coef_
```

```python
#making predictions
predicted=lin_model.predict(X_test)
```

```python
plt.scatter(X_test,Y_test,color='b')
plt.plot(X_test,predicted,color='r')
```

```python

p=pd.DataFrame(list(zip(X_test,Y_test,predicted)),columns=['x','target_y','predicted'])
p=p.round(decimals=2)
p.head()
```

```python
predicted=predicted.reshape(-1,1)
xtest=X_test.reshape(-1,1)
ytest=Y_test.reshape(-1,1)

xtest_scaled=x_scaler.inverse_transform(xtest)
ytest_scaled=y_scaler.inverse_transform(ytest)
predicted_scaled=y_scaler.inverse_transform(predicted)

xtest_scaled=xtest_scaled[:,-1]
ytest_scaled=ytest_scaled[:,-1]
predicted_scaled=predicted_scaled[:,-1]

p=pd.DataFrame(list(zip(xtest_scaled,ytest_scaled,predicted_scaled)),columns=['x','target_y','predicted'])
p=p.round(decimals=2)
p.head()
```

```python
mean_squared_error(ytest,predicted)
```
